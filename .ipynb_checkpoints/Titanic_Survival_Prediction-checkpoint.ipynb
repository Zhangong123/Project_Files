## 泰坦尼克号生存率预测模型
- 最终最高预测准确率：**79.186%**
- 准确率排名：前**7%**
- 步骤梳理：
    * 数据清洗：
        1. 对Ticket特征的字符串和票号进行分离，变成两列
        2. Embark数据存在极少量缺失值，填写众数对应类型
        3. 对Embark数据用one-hot-encoding技术分为3列特征（0/1）
        4. Sex数据替换成0/1
        5. 删除Name与Cabin（缺失值过多）特征，这里其实可以转化为tokens的形式来进行数据集的组建，考虑到数据集行数过少（过拟合风险），故删除
        6. Age数据缺失项过多，将age转为目标变量，用神经网络构建模型进行age的预测，拟合R^2约为`35%` 
     * 特征工程：
        1. 用随机森林模型对特征重要性进行排序，age的神经网络模型采用同样的办法，并删去后四位特征，在数据集过少的情况下减少过拟合
        2. 对几个特征进行Z-Score归一化操作，便于模型能够快速收敛
        3. 由于第一个采用了逻辑回归模型算法，所以对数据集进行了多项式拓展
     * 模型选择与调优：
        1. 这里选用了逻辑回归和随机森林模型
        2. 分出交叉验证数据集，对逻辑回归的多项式级数与正则化（L2）系数进行多模型拟合调优选择，确定级数为3，正则化系数0.05
        3. 随机森林模型进行多模型参数调优，最终选择最大深度: 10, 特征数量选择方法: 'sqrt', 树数量: 200
        4. 随机森林过拟合现象非常明显，主要还是数据量过少。最终验证的准确率两个模型基本相同
     * 备注说明：
        1. 代码未经最终整理与优化

## Titanic Survival Rate Prediction Model
- Highest prediction accuracy: **79.186%**
- Accuracy ranking: Top **7%**
- Summary:
    * Data Cleaning:
        1. Separation of string and ticket number into two columns.
        2. Filling few missing values in the Embark column with the mode of the data.
        3. The Embark data is split into 3 columns using one-hot-encoding technique (0/1).
        4. Replacement of Sex data with 0/1.
        5. Deletion of Name and Cabin (due to excessive missing values) features. Actually, the Name feature can be transformed into tokens to build the dataset. However, considering the small number of rows in the dataset (risk of overfitting), it is deleted.
        6. Considering large number of missing values in the Age data, converting Age into target variable and using a neural network model for age prediction, which yields R^2 equals approximately 35%.
    * Feature Engineering:
        1. Using random forest model to rank feature contribution. Applying the same approach to the above age neural network model, the same last four features were removed to reduce overfitting.
        2. Applying Z-Score normalization to several features for quick convergence of the model.
        3. Since logistic regression model algorithm was used, polynomial expansion was applied to the dataset.
    * Model Selection and Tuning:
        1. Logistic regression and random forest models were selected.
        2. A cross-validation dataset was created to perform model fitting and tuning for the polynomial series and regularization (L2) coefficient in logistic regression. The chosen series was 3, and the regularization coefficient was 0.05.
        3. Multiple model parameter tuning was conducted for the random forest model. The final selected parameters were: maximum depth: 10, feature selection method: ‘sqrt’, and tree number: 200.
        4. Overfitting was significant in the random forest model, mainly due to the small dataset. The final report accuracy of both models was almost equal.
    * Notes:
        1. The code has not been finalized and optimized.

## Data Feature Explanation

- PassengerID: Identification ID for passengers
- Survived：Ground truth whether this passenger suvived
- Pclass: Ticket class(1 = 1st, 2 = 2nd, 3 = 3rd)
- Name: Name
- Sex: Sex
- Age: Age
- SibSp: Number of siblings / spouses aboard the Titanic
- Parch: Number of parents / children aboard the Titanic
- Ticket: Ticket number
- Fare: Passenger fare
- Cabin: Cabin number
- Embarked: Port of Embarkation(C = Cherbourg, Q = Queenstown, S = Southampton)

```python
# Take a look at how this dataset looks like, set passengerID as first column
df_train = df_train.set_index('PassengerId')
df_train_backup = df_train.copy()
df_train.head()
```

## Data Preprocessing

```python
df_train.shape
#Inculuding 891 passengers, 12 features
```

```python
#Replace male and female with 1 and 0
df_train['Sex'] = df_train['Sex'].replace({'male': 1, 'female': 0})
```

```python
df_train.drop('Name', axis = 1, inplace = True)
```

```python
"""Ticket can record some hiden info about passengers, so needed to be dealt with, some ticket id include prefix plus number,
some are not, so next i'll seperate this column into two features, and deal with some blank error.
"""
df_train['Ticket'] = df_train['Ticket'].str.replace('STON/O 2.','STON/O2.',regex = True).replace('SC/AH Basle','SC/AHBasle', regex = True) 
df_train_ticket = df_train['Ticket'].str.split(' ', expand = True)
df_train_ticket.head()
```

```python
#Now modify the dataset
#Seperate number with prefix, get number
df_train_ticket.loc[:,1] = df_train_ticket.loc[:,1].fillna(value = np.nan)
switch_condition = df_train_ticket.loc[:,1].isnull()
temp_train_ticket_0 = df_train_ticket.loc[switch_condition, 0]
temp_train_ticket_1 = df_train_ticket.loc[switch_condition == False, 1]

df_train['Ticket_number_0'] = temp_train_ticket_0
df_train['Ticket_number_0'] = df_train['Ticket_number_0'].str.replace('LINE', '-1')
df_train['Ticket_number_0'] = df_train['Ticket_number_0'].apply(np.nan_to_num)

df_train['Ticket_number_1'] = temp_train_ticket_1
df_train['Ticket_number_1'] = df_train['Ticket_number_1'].str.replace('LINE', '-1')
df_train['Ticket_number_1'] = df_train['Ticket_number_1'].apply(np.nan_to_num)

df_train['Ticket_number'] = df_train['Ticket_number_0'].astype(float) + df_train['Ticket_number_1'].astype(float)

df_train.drop('Ticket_number_0', axis = 1, inplace = True)
df_train.drop('Ticket_number_1', axis = 1, inplace = True)

df_train.head()
```

```python
#Seperate number with prefix, get prefix on a new comlumn
temp_train_ticket_2 = df_train_ticket.loc[switch_condition == False, 0]
df_train['Ticket_prefix'] = temp_train_ticket_2
replace_condition = df_train.loc[:,'Ticket_number'] == -1
df_train.loc[replace_condition, 'Ticket_number'] = 0
df_train.loc[replace_condition, 'Ticket_prefix'] = 'LINE'
df_train.head()
```

```python
#Delete Cabin and Ticket(already dealt with)
df_train.drop('Ticket', axis = 1, inplace = True)
df_train.drop('Cabin', axis = 1, inplace = True)
df_train.head()
```

```python
#Deal with Embarked feature using one-hot-encoding technique
one_hot_encoding_df = pd.get_dummies(df_train['Embarked'])
df_train = pd.merge(df_train, one_hot_encoding_df, on = 'PassengerId')
df_train.rename(columns = {'C':'Embarked-C', 'Q':'Embarked-Q','S':'Embarked-S'}, inplace = True)
df_train.drop('Embarked', axis = 1, inplace = True)
df_train.head()
```

```python
#Tackle with null value
#Column Age has 177 null values
#Column Ticket_prefix has 661 null values
for i in df_train.loc[:]:
    
    print(df_train[i].isnull().sum())
```

```python
#There are some Embarked data also missing(just 2), fill out with the mode of these three number
df_embarked = df_train.loc[:,'Embarked-C'] + df_train.loc[:,'Embarked-Q'] + df_train.loc[:,'Embarked-S']
print(df_train.loc[:,'Embarked-C'].sum(), df_train.loc[:,'Embarked-Q'].sum(), df_train.loc[:,'Embarked-S'].sum())

df_train.loc[df_embarked == 0, 'Embarked-S'] = 1
```

```python
#visualization between Fare/TicketNo and Age

fig, axs = plt.subplots(2, 1, figsize=(8, 8))
axs[0].scatter(df_train_age_model.loc[:,'Age'], df_train_age_model.loc[:,'Fare'])
axs[0].set_title('Fare vs Age')

axs[1].scatter(df_train_age_model.loc[:,'Age'], df_train_age_model.loc[:,'Ticket_number'])
axs[1].set_title('Ticket Number vs Age')
plt.tight_layout()
```

```python
#visualization between Ticket Prefix/Embarked Info and Age

def transform_to_num(a):
    if type(a) == str:
        return 1
    else:
        return 0

df_train_embarked = df_train_backup.loc[age_na_condition == False,'Embarked'].replace({'S':1, 'Q':2, 'C':3})

df_train_age_model.loc[:,'Ticket_prefix'] = df_train_age_model.loc[:,'Ticket_prefix'].apply(transform_to_num)

fig, axs = plt.subplots(2, 1, figsize=(8, 3))
axs[0].scatter(df_train_age_model.loc[:,'Age'], df_train_age_model.loc[:,'Ticket_prefix'])
axs[0].set_title('Ticket Prefix vs Age')

plt.yticks([0,1,2,3])
axs[1].scatter(df_train_age_model.loc[:,'Age'], df_train_embarked)
axs[1].set_title('Embarked vs Age')
plt.tight_layout()
```

```python
#Z-score Normalization
scaler = StandardScaler()
features_to_normalize = ['Fare', 'Ticket_number']
df_train_age_model[features_to_normalize] = scaler.fit_transform(df_train_age_model[features_to_normalize])

df_train_age_model.head()
```

```python
#After getting intuition about this data, use random forest to get high info gain features

rf = RandomForestRegressor(random_state = 12)
df_train_age_model_X = df_train_age_model.drop('Age',axis = 1).drop('Survived',axis = 1)
df_train_age_model_Y = df_train_age_model.loc[:,'Age']

rf.fit(df_train_age_model_X, df_train_age_model_Y)

importance = rf.feature_importances_

feature_importance_dict = dict(zip(df_train_age_model_X.columns, importance))

sorted_features = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)

for feature, score in sorted_features:
    print("Feature", feature, ":", score)
```

```python
#Drop last four columns
columns_to_drop = ['Ticket_prefix', 'Embarked-C', 'Embarked-Q','Embarked-S']
df_train_age_model_X.drop(columns_to_drop, axis=1, inplace=True)

X_train, y_train = df_train_age_model_X, df_train_age_model_Y
```

```python
#Build a neural network to estimate 177 mising age values
units_list = [40, 20,10, 1]
age_model = Sequential([
Dense(units = units_list[0], activation = 'relu'),
Dense(units = units_list[1], activation = 'relu'),
Dense(units = units_list[2], activation = 'relu'),
Dense(units = units_list[3], activation = 'linear')
])

age_model.compile(optimizer='adam', loss='mean_squared_error')

age_model.fit(X_train, y_train, epochs = 400)
```

```python
#Fill missing values in Age, using trained model
df_age_prediction = df_train.loc[age_na_condition == True,:]

#Normalization using same old scaler
df_age_prediction[features_to_normalize] = scaler.fit_transform(df_age_prediction[features_to_normalize])
df_age_prediction.loc[:,'Ticket_prefix'] = df_age_prediction.loc[:,'Ticket_prefix'].apply(transform_to_num)

df_age_prediction = df_age_prediction.drop(columns_to_drop, axis=1).drop('Survived', axis=1)
df_age_prediction
```

```python
#Feature selection
rf_allset = RandomForestClassifier(random_state = 12)
df_train_X = df_train.drop('Survived',axis = 1)
df_train_Y = df_train.loc[:,'Survived']

rf_allset.fit(df_train_X, df_train_Y)

importance_allset = rf_allset.feature_importances_

feature_importance_dict_allset = dict(zip(df_train_X.columns, importance_allset))

sorted_features_allset = sorted(feature_importance_dict_allset.items(), key=lambda x: x[1], reverse=True)

for feature, score in sorted_features_allset:
    print("Feature", feature, ":", score)
```

```python
#Drop last four columns
columns_to_drop_allset = ['Ticket_prefix', 'Embarked-C', 'Embarked-Q','Embarked-S']
df_train_X.drop(columns_to_drop_allset, axis=1, inplace=True)
```

```python
#Z-score Normalization
scaler_allset = StandardScaler()
features_to_normalize = ['Pclass','SibSp','Parch','Age','Fare', 'Ticket_number']
df_train_X[features_to_normalize] = scaler_allset.fit_transform(df_train_X[features_to_normalize])
```

```python
#Fit to logistic regression
train_accuracy=list()
test_accuracy=list()

for index, degree in enumerate(range(1,10)):
    poly = PolynomialFeatures(degree)
    df_train_X_poly = poly.fit_transform(df_train_X)

    X_train_allset, X_test_allset, y_train_allset, y_test_allset = train_test_split(df_train_X_poly, df_train_Y, test_size=0.2, random_state=1)

    logistic_model = LogisticRegression(penalty = 'l2', C=0.05)
    logistic_model.fit(X_train_allset, y_train_allset)
    #Evaluation
    train_prediciton = logistic_model.predict(X_train_allset)
    test_prediction = logistic_model.predict(X_test_allset)
    train_accuracy.append(accuracy_score(y_train_allset, train_prediciton))
    test_accuracy.append(accuracy_score(y_test_allset, test_prediction))

plt.scatter([range(1,10)], train_accuracy, label = 'train')
plt.scatter([range(1,10)], test_accuracy, label = 'test')
plt.legend()
plt.show()

#So we choose degree to be 3 or 4
```

```python
#Now choose regularization parameter
train_accuracy=list()
test_accuracy=list()
Cvalue_list=list()

Cvalue_x = 0.0001
step = 0.0001

poly = PolynomialFeatures(3)
df_train_X_poly = poly.fit_transform(df_train_X)
X_train_allset, X_test_allset, y_train_allset, y_test_allset = train_test_split(df_train_X_poly, df_train_Y, test_size=0.2, random_state=1)

for index, Cvalue in enumerate(range(1,15)):
    Cvalue_list.append(Cvalue_x)
    logistic_model = LogisticRegression(penalty = 'l2', C=Cvalue_x)
    logistic_model.fit(X_train_allset, y_train_allset)
    

    train_prediciton = logistic_model.predict(X_train_allset)
    test_prediction = logistic_model.predict(X_test_allset)
    train_accuracy.append(accuracy_score(y_train_allset, train_prediciton))
    test_accuracy.append(accuracy_score(y_test_allset, test_prediction))
    
    step *= 2
    Cvalue_x += step

plt.scatter(Cvalue_list, train_accuracy, label = 'train')
plt.scatter(Cvalue_list, test_accuracy, label = 'test')
plt.legend()
plt.show()

print(Cvalue_list)
#So lamda to be 0.05
```

```python
#Train our final model

poly = PolynomialFeatures(3)
df_train_X_poly = poly.fit_transform(df_train_X)

X_train_allset, y_train_allset = df_train_X_poly, df_train_Y

logistic_model = LogisticRegression(penalty = 'l2', C=0.05)
logistic_model.fit(X_train_allset, y_train_allset)

train_prediciton = logistic_model.predict(X_train_allset)
#test_prediction = logistic_model.predict(X_test_allset)
train_accuracy = accuracy_score(y_train_allset, train_prediciton)
#test_accuracy = accuracy_score(y_test_allset, test_prediction)

train_accuracy#, test_accuracy
```

```python
#Try again using Random Forest Algorithm, find best parameters
X_train_allset,y_train_allset = df_train_X, df_train_Y
rf_run = RandomForestClassifier(random_state = 12)

param_grid = {
    'n_estimators': [30, 50, 100, 200, 300],
    'max_depth': [None,3,5,7,8,9,10],
    'max_features': ['sqrt', 'log2'],
}

grid_search = GridSearchCV(estimator=rf_run, param_grid=param_grid, cv=5)

grid_search.fit(X_train_allset, y_train_allset)

print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```
